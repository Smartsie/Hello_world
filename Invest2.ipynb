```
# Chaque action a un return
interet=[1.0266246831647217,1.0386802046259886]
variation_interet_par_an=[0.05830390458402494,0.05287636012575688]
```

```
def generate_interest2(interest, variation,year=10):
  profile=[interest]
  for i in range(0,year):
    #interest+=(np.random.random(1)-0.5)*variation
    interest+=np.random.normal(0,variation)
    profile.append(interest)
  return profile
```

```
def generate_interest(interest,variation, year):
  x=np.random.normal(interest, variation,year)
  return np.cumprod(x)
```

```
np.cumprod([1,2,3])
```

```
years=10
results_95=[]
All_traces=[]
for a,b in zip(interet,variation_interet_par_an):
  # generate N profiles for that interest
  N=1000
  profile=[]
  summary=[]
  for j in range(0,N):
    trace=generate_interest(a,b,years)
    cash_trace=[1]
    for i in range(1,years):
      cash_trace.append(cash_trace[-1]*(1+trace[i]))
    profile.append(cash_trace)
    plt.plot(profile[j],color='blue',alpha=0.05)
    summary.append(cash_trace[-1])
  All_traces.append(profile)
  results_95.append(np.percentile(summary,[2.5,97.5]))
  plt.show()
```

```
Cash=10000
```

```
# What is best combination of portfolio which will minimize risk and maximize profit
experiments=5
a=np.linspace(0,1,experiments)
b=1-a
a=a*Cash
b=b*Cash
```

```
N=50
z=[]

for C1,C2 in zip(a,b):
  perc=[]
  for j in range(0,N):
    trace_a=np.multiply(C1,generate_interest(interet[0],variation_interet_par_an[0],10))
    trace_b=np.multiply(C2,generate_interest(interet[1],variation_interet_par_an[1],10))
    total=trace_a+trace_b
    plt.plot(total, color='blue',alpha=0.1)
    perc.append(total[-1])
  z.append([C1,C2,np.percentile(perc,50),np.percentile(perc,5),np.percentile(perc,95)])
plt.show()
```

```
N=500
z=[]

for C1,C2 in zip(a,b):
  perc=[]
  for j in range(0,N):
    trace_a=np.multiply(C1,All_traces[0][np.random.randint(len(All_traces[0]))])
    trace_b=np.multiply(C2,All_traces[1][np.random.randint(len(All_traces[0]))])
    total=trace_a+trace_b
    plt.plot(total, color='blue',alpha=0.1)
    perc.append(total[-1])
  z.append([C1,C2,np.percentile(perc,50),np.percentile(perc,5),np.percentile(perc,95)])
plt.show()
```
